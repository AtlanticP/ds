```python
import scipy.stats as sts
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def distr(val):
    
    if val <= 0.2:
        return 0
    elif val >= 0.2 and val < 0.5:
        return 10
    elif val >= 0.5 and val < 0.55:
        return 15
    elif val >= 0.55 and val < 0.90:
        return 25
    else:
        return 30
```

```python
D = {0: 0.2, 10:0.3, 15:0.05, 25:0.35, 30:0.1}
print('Distribution: ', D)

E = sum([x*p for x, p in D.items()])
print('Math expectation: ', E)

d = sum([(x-E)**2*p for x, p in D.items()])
print('Dispersion: ', d)
print(np.sqrt(d/30))

# D_inv = {p: x for x, p in D.items()}
```

```python
means = []
var = []
for i in range(1, 1000):
    yy = np.array([distr(i) for i in np.random.rand(30)])
    means.append(yy.mean()), var.append(yy.var().astype(int))
    print(f'{i} Var: ', yy.var())
    print(f'{i} Mean: ', yy.mean())
    print()
```
